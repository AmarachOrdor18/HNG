```python
# Combine all DataFrames by stacking them vertically
merged_df = pd.concat([Telex, Mattermost, Slack], ignore_index=True)
```

```python
# Display basic information
print(merged_df.info())

# Display the first few rows
print(merged_df.head())

# Check for missing values
print(merged_df.isna().sum())

# Summary statistics
print(merged_df.duplicated().sum())
```

```python
merged_df = merged_df.drop_duplicates()
```

```python
# Display basic information
print(merged_df.info())

# Display the first few rows
print(merged_df.head())

# Check for missing values
print(merged_df.isna().sum())

# Summary statistics
print(merged_df.duplicated().sum())
```

```python
merged_df = merged_df.drop(columns=["Top of page bid (low range)", "Top of page bid (high range)"])
```

```python
# Display basic information
print(merged_df.info())

# Display the first few rows
print(merged_df.head())

# Check for missing values
print(merged_df.isna().sum())

# Summary statistics
print(merged_df.duplicated().sum())
```

```python
merged_df["Competition (indexed value)"] = merged_df["Competition (indexed value)"].fillna(merged_df["Competition (indexed value)"].mean())
```

```python
# Display basic information
print(merged_df.info())

# Display the first few rows
print(merged_df.head())

# Check for missing values
print(merged_df.isna().sum())

# Summary statistics
print(merged_df.duplicated().sum())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


# Vectorize keywords using TF-IDF
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(merged_df["Keyword"])

# Determine the optimal number of clusters using the Silhouette Score
best_clusters = 3  # Default cluster count
best_score = -1

for k in range(2, 6):  # Trying different cluster sizes (removed research)
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    
    if score > best_score:
        best_score = score
        best_clusters = k

print(f"Optimal number of clusters: {best_clusters}")

# Apply KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=best_clusters, random_state=42, n_init=10)
merged_df["cluster"] = kmeans.fit_predict(X)

# Define intent labels (Removed "Research")
intent_labels = {
    0: "Informational",
    1: "Navigational",
    2: "Transactional",
    3: "Commercial"
}

# Ensure every keyword is categorized correctly
merged_df["intent"] = merged_df["cluster"].map(lambda x: intent_labels.get(x, "Commercial"))  # Default to "Commercial"
```

```python
# Display basic information
print(merged_df.info())

# Display the first few rows
print(merged_df.head())

# Check for missing values
print(merged_df.isna().sum())

# Summary statistics
print(merged_df.duplicated().sum())
```

```python
merged_df.columns
```

```python
# Convert necessary columns to numeric (in case they are strings)
merged_df["Avg. monthly searches"] = pd.to_numeric(merged_df["Avg. monthly searches"], errors="coerce")
merged_df["Three month change"] = pd.to_numeric(merged_df["Three month change"], errors="coerce")
merged_df["YoY change"] = pd.to_numeric(merged_df["YoY change"], errors="coerce")

# Drop rows where "Avg. monthly searches" < 500 AND both "Three month change" and "YoY change" <= 0
df_filtered = merged_df[
    (merged_df["Avg. monthly searches"] >= 500) | 
    (merged_df["Three month change"] > 0) | 
    (merged_df["YoY change"] > 0)
]
```

```python
df_filtered.shape
```

```python
import matplotlib.pyplot as plt

# Aggregate average monthly searches by intent
intent_avg_searches = df_filtered.groupby("intent")["Avg. monthly searches"].sum()

# Plot bar chart
plt.figure(figsize=(8, 5))
intent_avg_searches.plot(kind="bar", color=["blue", "green", "orange", "red"])
plt.title("Total Search Volume by Keyword Intent")
plt.xlabel("Intent")
plt.ylabel("Avg. Monthly Searches")
plt.xticks(rotation=45)
plt.show()
```

```python
# Sort by "Avg. monthly searches" for better visualization
df_sorted = df_filtered.sort_values("Avg. monthly searches", ascending=False).head(20)

plt.figure(figsize=(10, 5))
plt.plot(df_sorted["Keyword"], df_sorted["Three month change"], marker="o", label="Three Month Change", color="blue")
plt.plot(df_sorted["Keyword"], df_sorted["YoY change"], marker="s", label="YoY Change", color="green")

plt.xticks(rotation=75)
plt.title("Keyword Growth Trends (3-Month & YoY Change)")
plt.xlabel("Keywords")
plt.ylabel("Change Percentage")
plt.legend()
plt.grid()
plt.show()
```

```python
# Get the top 5 keywords with the highest average monthly searches
top_keywords = df_filtered.nlargest(5, "Avg. monthly searches")[["Keyword", "Avg. monthly searches"]]

# Display the top 5 keywords
print("Top 5 Keywords by Avg. Monthly Searches:")
print(top_keywords)

# Plot bar chart for the top 5 keywords
plt.figure(figsize=(8, 5))
plt.barh(top_keywords["Keyword"], top_keywords["Avg. monthly searches"], color="purple")
plt.xlabel("Avg. Monthly Searches")
plt.ylabel("Keyword")
plt.title("Top 5 Keywords by Avg. Monthly Searches")
plt.gca().invert_yaxis()  # Highest value on top
plt.show()
```

```python
# Save results to a new CSV file
df_filtered.to_csv("DataCleaned_keyword_cluster.csv", index=False)

print("Keyword clustering completed! Results saved in 'DataCleaned_keyword_cluster.csv'")
```
